```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def categorizar(whiteWolf):
    paradas=[0.5,0.7,0.9,0.95,0.99]

    listaQuartiles=[]
    for x in range(len(whiteWolf)):

        if (whiteWolf['prediccion'][x]==0):
            #print(whiteWolf['prediccion'][x])
            f=0
            #print(f)
            listaQuartiles.append(f)
        elif(whiteWolf['prediccion'][x]<paradas[0]):
            #print(whiteWolf['prediccion'][x])
            f=1
            #print(f)
            listaQuartiles.append(f)

        elif(whiteWolf['prediccion'][x]<paradas[1]):
            #print(whiteWolf['prediccion'][x])
            f=2
            #print(f)
            listaQuartiles.append(f)

        elif(whiteWolf['prediccion'][x]<paradas[2]):
            #print(whiteWolf['prediccion'][x])
            f=3
            #print(f)
            listaQuartiles.append(f)

        elif(whiteWolf['prediccion'][x]<paradas[3]):
            #print(whiteWolf['prediccion'][x])
            f=4
            #print(f)
            listaQuartiles.append(f)

        elif(whiteWolf['prediccion'][x]<paradas[4]):
            #print(whiteWolf['prediccion'][x])
            f=5
            #print(f)
            listaQuartiles.append(f)

        elif(whiteWolf['prediccion'][x]<1):
            #print(whiteWolf['prediccion'][x])
            f=6
            #print(f)
            listaQuartiles.append(f)

    whiteWolf['prediccion']=listaQuartiles
    return whiteWolf




def heatmapMatriz(texto,categorizado=False):
    
    direccion=('DFlistoParaHeatMap/'+texto+'.csv')
    whiteWolf = pd.read_csv(direccion, index_col=0)
    
    print(f'\nThe chances of fire in all the area is {whiteWolf.prediccion.mean()}')
    
    if categorizado==True:
        whiteWolf=categorizar(whiteWolf)
    
    yosemite=[]
    for x in range(len(whiteWolf)):
        
        yosemite.append(whiteWolf['prediccion'][x])
        #yosemite.append(whiteWolf['prediccion'][x])
    
    n=len(yosemite)
    ns = np.ceil(np.sqrt(n)).astype(int)
    whiteWolf = np.reshape(yosemite,(ns,ns))

    # Fixing random state for reproducibility
    plt.suptitle(texto)

    plt.imshow(whiteWolf, cmap=plt.cm.YlOrRd)

    #niveles=['a','b','c','d','e','f']
    plt.subplots_adjust(bottom=0.1, right=0.8, top=0.9)
    cax = plt.axes([0.85, 0.1, 0.075, 0.8])
    plt.colorbar(cax=cax)
    plt.show()
```

```python
direccion='2019-02-22EN-119.9_37.75A30.0kmXlado60puntosXkmET'
df=heatmapMatriz(direccion,True)
```
